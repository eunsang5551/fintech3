# 전체 데이터 수집

```python
import os
import re
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup as bs
from dotenv import load_dotenv
load_dotenv()
```

```python
start = 1
end = 1000
page = 1
total_page = 1
result = {}
while page <= total_page:
    url = f"http://openapi.seoul.go.kr:8088/sample/json/VwsmAdstrdNcmCnsmpW/1/5/"
    r = requests.get(url)
#     print(r.url)
#     print(r.status_code)
    response = r.json()
    if total_page ==1:
        total_page = response["VwsmAdstrdNcmCnsmpW"]['list_total_count'] // 1000 + 1
        
    kor_cols = ["기준_년분기_코드", "행정동_코드", "행정동_코드_명", "월_평균_소득_금액",
                "소득_구간_코드", "지출_총금액", "식료품_지출_총금액", "의류_신발_지출_총금액",
                "생활용품_지출_총금액", "의료비_지출_총금액", "교통_지출_총금액",
                "교육_지출_총금액", "유흥_지출_총금액", "여가_문화_지출_총금액",
                "기타_지출_총금액", "음식_지출_총금액"]
    
    for item in response["VwsmAdstrdNcmCnsmpW"]['row']:
        for key, value in zip(kor_cols, item.values()):
            if type(value) == float:
                result.setdefault(key, []).append(int(value))
            else:
                result.setdefault(key, []).append(value)
    
    print(f"{page}/{total_page}데이터 수집중", end="\r")
#     print(f"{start}, {end}")
    start += 1000
    end += 1000
    page += 1
    
df = pd.DataFrame(result)
df
```
