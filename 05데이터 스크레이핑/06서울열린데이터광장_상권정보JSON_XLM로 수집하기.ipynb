```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup as bs
from dotenv import load_dotenv
import os
import re
load_dotenv()
```

```python
start = 1
end = 1000
page = 1
total_page = 1
result = {}

while page <= total_page:
    url =f"http://openapi.seoul.go.kr:8088/sample/json/VwsmAdstrdNcmCnsmpW/1/5/"
    r = requests.get(url)
    print(r.status_code)
    response = r.json()
    if total_page == 1:
        total_page = response['VwsmAdstrdNcmCnsmpW']['list_total_count'] // 1000 + 1


    for item in response['VwsmAdstrdNcmCnsmpW']['row']:
        for key, value in item.items():
            if type(value) == float:
                result.setdefault(key, []).append(int(value))
            else:
                result.setdefault(key, []).append(value)

    print(f"{page}/{total_page} 데이터 수집중", end="\r")            
    start += 1000
    end += 1000
    page += 1
    
    time.sleep(0.2)
            
df = pd.DataFrame(result)
df
```

```python
start = 1
end = 1000
page = 1
total_page = 1
result2 = {}

while page <= total_page:
    url =f"http://openapi.seoul.go.kr:8088/sample/json/VwsmAdstrdNcmCnsmpW/1/5/"
    r = requests.get(url)
#     print(r.status_code)
    response = bs(r.text, 'xml')
    total_page = int(response.select_one("list_total_count").text) // 1000 + 1
    tag_list = [tags.name for tags in response.select_one('row').find_all()]

    kor_cols = ["기준_년분기_코드", "행정동_코드", "행정동_코드_명", "월_평균_소득_금액", "소득_구간_코드",
                "지출_총금액", "식료품_지출_총금액", "의류_신발_지출_총금액", "생활용품_지출_총금액",
                "의료비_지출_총금액", "교통_지출_총금액", "교육_지출_총금액", "유흥_지출_총금액",
                "여가_문화_지출_총금액", "기타_지출_총금액", "음식_지출_총금액"]
    
    for tag, kor_col in zip(tag_list, kor_cols):
        for value in response.select(tag):
            result2.setdefault(kor_col, []).append(value.text)

    print(f"{page}/{total_page} 데이터 수집중", end="\r")            
    start += 1000
    end += 1000
    page += 1

    time.sleep(0.2)
        
df2 = pd.DataFrame(result2)
df2
```
