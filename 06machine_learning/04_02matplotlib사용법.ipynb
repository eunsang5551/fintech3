# matplotlib의 2가지 코딩 스타일

## 00-스타일: matplotlib의 공식 추천 스타일
* 복잡한 그래프를 그릴 때 주로 사용
* 사용법이 pyplot 스타일보다 복잡함

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig,ax = plt.subplots(figsize=(5,2.7))
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('간단 그래프')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(5, 7))
ax[0].plot(x, x, label='linear')
ax[1].plot(x, x, label='quadratic')
ax[2].plot(x, x, label='cubic')
ax[0].set_xlabel('x label')
ax[0].set_ylabel('y label')
ax[0].set_title('간단 그래프')
ax[1].set_xlabel('x label')
ax[1].set_ylabel('y label')
ax[1].set_title('quadratic 간단 그래프')
ax[2].set_xlabel('x label')
ax[2].set_ylabel('y label')
ax[2].set_title('cubic 간단 그래프')
plt.tight_layout()
plt.show()
```

# pyplot 스타일

# 선 모양 바꾸기 set_linestyle(모양)

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
```

```python
data1
```

```python
data2
```

```python
len(data1)
```

선모양: '-', '--', '-.', ':'

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle=":")
ax.plot(x, np.cumsum(data2), color='orange', linewidth=2, linestyle="-.")
plt.show()
```

# 산점도 그래프를 그리고 색 변경하기
* c(안쪽색), edgecolors(테두리색)

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1, data2, s=50, c='c', edgecolors='face')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1, data2, s=50, c='c', edgecolors='k')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter(data1, data2, s=50, c=(1,0,0), edgecolors=('k',0.5))
plt.show()
```

# 마커의 모양 지정하기

```python
fig, ax = plt.subplots(figsize=(5,4))
ax.plot(data1, '8', label='data1')
ax.plot(data2, 'D', label='data2')
ax.plot(data3, '3', label='data3')
ax.plot(data4, '>', label='data4')
ax.legend()
plt.show()
```

# x축, y축 이름, 눈금, 그래프이름

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7))
n, bins, patches = ax.hist(x, 50, density=True, color='C0', alpha=0.75)
ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title("Aardvark lengths \n (not really)")
ax.text(75, 0.025, r'$\mu=115, \ \sigma=15$')
ax.axis([60, 160, 0, 0.03]) 
plt.show()
```

# 눈금자 단위 구간 설정하기

```python
xdata = np.arange(len(data1))
fig, ax = plt.subplots(2, 1, figsize=(5, 2.7), layout='constrained')
ax[0].plot(xdata, data1)
ax[0].set_title("automatic ticks")

ax[1].plot(xdata, data1)
ax[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
ax[1].set_yticks([-1.5, 0, 1.5])
ax[1].set_title("manual ticks")
plt.show()
```
