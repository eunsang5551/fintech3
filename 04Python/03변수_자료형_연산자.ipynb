# 자료형(data type)
## 파이썬의 자료형은 동적으로 변화
* int: 정수(소수점이 없는 숫자)
* float: 실수(소수점이 있는 숫자)
* str: 문자 / 문자열
* list: 여러개의 자료가 순서대로 저장된 자료형, 모든 자료형을 넣어 만들 수 있음 [자료]
* tuple: list와 비슷하나 한 번 만들면 수정x, 모든 자료형을 넣어 만들 수 있음 (자료)
* dictionary(dict): key:value 쌍으로 이루어진 자료형, 모든 자료형을 넣어 만들 수 있음 {key:value}
* set: 집합자료형, 자료의 중복이 되지 않음 {자료}
