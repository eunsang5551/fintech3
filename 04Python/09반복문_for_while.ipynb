# 반복문 for, while
* 반복문은 동일한 작업을 여러번 반복해야 할 경우 사용
* for, while 두종류의 반복문이 있음
* for는 **반복 횟수가 정해져 있는 경우** 사용
* while 반복 횟수가 정해져 있지 않은 경우 사용(무한반복)

# for 반복문: 반복할 횟수가 정해져 있는 경우
for 변수1, 변수2 in 반복할 객체/자료(문자열, 리스트, 튜플, 딕셔너리):<br>
____반복할 코드

```python
# 숫자 생성 함수 range(시작, 끝, step)
```

```python
# for문을 이용해서 숫자 1-10출력
for num in range(1, 11):
    print(num, end=" ")
```

# countinue / break
* countinue: 코드 실행을 건너 뛰고 반복 계속 진행
* break: 특정 조건이 되었을 때 반복문 정지

```python
for idx, i in enumerate(s):
    if idx % 2 == 0:
        continue
        print(idx, i)
    else:
        if idx > 19:
            break
        print(idx,i)
```

# while: 반복할 횟수가 정해지지 않은 경우
* 무한 반복 가능
* 무한 반복이 되지 않도록 조건식을 잘 짜주는 것이 중요

변수 선언 및 초기화 num = 0<br>
while 조건식:<br>
____반복할 코드<br>
____변수 증가/감소 코드<br>
____반복문이 종료되는 조건 + break

* while 문으로 1-10까지 출력하기

```python
i = 1
while i <= 10:
    print(i, end =" ")
    i += 1
```

```python
i = 1
while True:
    print(i, end =" ")
    if i >= 10:
        break
    i += 1
```

* 1-50까지 수 중에서 3의 배수이거나 짝수인 수를 찾아서 합계를 출력하세요.

```python
i = 1
hap = 0
while i <= 50:
    if i % 3 == 0 or i % 2 == 0:
        hap += i
        print(f"i: {i}, hap: {hap}")
```

* 반복문 3개로 시계 만들기
* 초: 00-59
* 분: 00-59
* 시: 00-23
import time
time.sleep(0.1)
* 00:00:00 - 23:59:59

```python
import time
for si in range(0, 24):
    for bun in range(0, 60):
        for cho in range(0, 60):
            print(f"{si:02d}:{bun:02d}:{cho:02d}", end = "\r")
            time.sleep(1)
```
