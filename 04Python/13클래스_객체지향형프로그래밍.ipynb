```python
class NumberList():
    def __init__(self, numbers):
        self.numbers= numbers
    def squared(self):
        return [x**2 for x in self.numbers]
```

```python
squared = NumberList([1,2,3,4])
squared.squared()
```

```python
class FourCal():
    def setdata(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print(f"self.num1: {self.num1}")
        print(f"self.num2: {self.num2}")
              
    # 덧셈
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
    
    def mul(self):
        result = self.num1 * self.num2
        return result
    
    def div(self):
        result = self.num1 / self.num2
        return result  
```

```python
class Date():
    def __init__(self, year, month, day):
        self.year, self.month, self.day = year, month, day
    
    def print_date(self):
        print(self.year, self.month, self.day)
    
    @classmethod
    def from_string(cls, date_str):
        year, month, day = map(int, date_str.split("-"))
        return cls(year, month, day)
```

```python
dul = Date(2025, 10, 2)
```

```python
dul.print_date()
```

```python
dul2 = Date.from_string("2025-10-03")
```

```python
dul2.print_date()
```

```python
from abc import *
```

```python
# 추상 클래스 
class CoffeeBase(metaclass=ABCMeta):
    @abstractmethod
    def bean_input(self):
        pass
    
    @abstractmethod
    def grind(self):
        pass
```

```python
class CoffeeMachine(CoffeeBase):
    def bean_input(self):
        print("커피 원두 넣기")
    
    
    def grind(self):
        print("커피 원두 갈기")
```

```python
sam = CoffeeMachine()
```

```python
sam.bean_input()
```

```python
sam.grind()
```
