# 문자열과 문자열 함수

## 인덱싱과 슬라이싱
* 문자: 글자 1개
* 문자열: 글자가 여러개 => 순서가 있다. => 앞에서부터 0, 1, 2.. 순으로 번호가 매겨짐
* 순서가 있는 자료는 번호를 불러서 그 번호에 해당하는 자료를 꺼낼 수 있다.

* 인덱싱 변수명 [인덱스번호] 로 자료 1개를 꺼내는 것

```python
h1 = "문자열은 왼쪽부터 번호를 매길 수 있으며 시작은 0번부터 입니다."
print(h1)
```

```python
# h1변수에서 인덱스 10번의 자료를 추출하세요.
h1[10]
```

* 슬라이싱 변수명 [시작인덱스:끝인덱스+1: 간격]
* 시작인덱스와 끝 인덱스는 생략 가능
    * [:10] => 시작부터 9번인덱스까지 슬라이스
    * [5:] => 인덱스 5번부터 시작해서 끝까지 슬라이스
    * [:] => 전체를 다 가져와라.
    * [::2] => 전체 자료에서 2칸 씩 건너 뛰면서 슬라이스
    * [::-1] => 전체 자료를 역순 정렬

```python
# h1에서 "번호를"만 추출하세요
h1[10:13]
```

```python
# 인덱스 0번부터 15번까지 추출하세여
h1[0:16]
```

```python
# 인덱스 15번부터 끝까지 추출하세요
h1[15:]
```

```python
# 인덱스 1번부터 끝까지 2칸 씩 건너 뛰면서 추출하세요
h1[1::2]
```

```python
# h1의 자료를 역순 정렬하세요
h1[::-1]
```

# 문자열 함수
* replace(): 값을 찾아서 바꾸는 함수
* split(): 공백을 기준으로 자료를 나눠서 **리스트로 변형**
* join(): 기준자를 통해서 리스트를 하나의 **문자열로 변형**
* strip(): 문자열 외부의 공백 제거, 공백+특수문자 제거

## 문자열 바꾸기 replace(찾을 값, 바꿀 값, 반복횟수-1)
* replace 함수는 사용 후 결과는 재할당 해야 함.

```python
h2 = "He!ll!o Python!"
h2
```

```python
h2.replace("!", "")
```

```python
h2 = h2.replace("!", "", 2)
h2
```

```python
h3 = "He!ll!o# Pyt#hon!"
h3
```

## 함수의 체인메서드 함수를 . 으로 연결해 중첩사용

```python
# h3에서 !와 #을 없애고 마지막 ! 는 남겨두세요.
h3 = h3.replace("!", "", 2).replace("#", "")
h3
```

# strip()
* 문자열의 양쪽 공백과 특정문자 제거
* lstrip(): 왼쪽 공백 삭제
* rstrip(): 오른쪽 공백 삭제
* strip(): 양쪽 공백 삭제

```python
h6 = "    공백이 있는 문자열    "
h6
```

```python
h6.lstrip()
```

```python
h6.rstrip()
```

```python
h6.strip()
```

# 정규표현식
* 정규표현식: 문자의 패턴을 찾아서 치환하거나 그 패턴에 맞는 것만 추출
* import re
* re.sub(r'[^가-힣a-zA-Z0-9]', " ", 처리할 문자변수)

```python
import re
```

```python
text = "안녕하세요!~ This is a test.  1234 #$%^()"
```

```python
re.sub(r'[^가-힣a-zA-Z0-9]', " ", text).replace("   ", " ").strip()
```

```python
test3 = re.sub(r'[^가-힣a-zA-Z0-9]', " ", test3).replace("  ", " ").replace("  ", " ").strip().replace("1 5리터", "1.5리터")
```
