# 리스트와 튜플 사용하기 초급편

## 리스트 만들기
* 대괄호 안에 자료를 나열해서 만듬
* 내용이 비어있는 리스트도 생성 가능 []
* 자료가 1개만 있는 리스트도 생성 가능
* 숫자로만 된 리스트의 경우 range()와 list 형변환 함수로 생성 가능

```python
# 빈 리스트 만들기
l1 = []
l1
```

```python
print(l1, len(l1), type(l1))
```

```python
# 대괄호 안에 자료를 나열해서 만들기
l2 = [1,2,3, 3.5, "리스트", (4,5,6), {"key": "value"}]
```

```python
print(l2, type(l2))
```

## range 함수로 리스트 만들기
* range(시작번호, 끝번호+1, step)

```python
list(range(1, 11, 2))
```

```python
# 4의 배수로 리스트 만들기
list(range(4, 101, 4))
```

```python
# 역순으로 리스트 만들기
list(range(100, 1, -4))
```

## 리스트의 자료 삭제하기
* del 변수명[인덱스 번호]: 해당 인덱스의 자료가 삭제
* pop(): 리스트의 가장 끝 자료가 출력되면서 삭제됨
* remove(값): 리스트의 앞에서부터 입력한 값과 일치하는 값을 찾아서 삭제(정수, 실수 구분x)
