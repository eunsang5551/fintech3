# dict 함수를 이용해서 dict 만들기
* dict(key1 = value1, key2 = value2, key3 = value3)
* key 부분에 ""(따옴표)를 사용하지 않는다.
* key는 모두 문자형으로 변환된다.
* value가 문자형이면 반드시 ""(따옴표)가 필요하다.

```python
dict_ex1 = dict(정수=100, 실수=3.5, 문자="문자열",
               리스트=[1,2,3], 튜플=(4,5,6),
               딕셔너리={'key': "value"})
dict_ex1
```

* dict([(key1, value1), (key2, value2), (key3, value3)])
* 리스트 안에 key, value 쌍으로 된 튜플을 넣어서 만듬
* key는 자료형에 맞춰서 작성

```python
dict_ex1['실수']
```

```python
dict_ex2 = dict([(1, 10), (3.5, "3.5"), ("문자열", "문자열입니다."),
                ("list", [1,2,3]), ("tuple", (4,5,6)),
                ("dict", {"key" : "value"})])
dict_ex2
```

```python
dict_ex2[1]
```

```python
dict_ex2[3.5]
```

* dict(zip([key1, key2, key3], [value1, value2, value3]))
* key 묶음 리스트와 value 묶음 리스트를 zip함수로 합친 후 dict 생성
* key 묶음의 요소 개수와 value 묶음의 요소 개수가 일치해야함

```python
# zip함수 예시
for i in [1,2,3]:
    print(i, end=" ")
```

```python
for i in zip([1,2,3], [4,5,6], [7,8]):
    print(i, end=" ")
```

```python
dict_ex3 = dict(zip([100, 3.5, '문자열'], [10, 12.5, "문자열입니다."]))
dict_ex3
```

```python
dict_ex3[100]
```

# 딕셔너리 함수 사용하기

## setdefault(key, value) 딕셔너리에 자료 추가
* 딕셔너리에 key와 value를 추가
* key가 중복되면 나중에 추가하는 값을 무시함(원래 자료 보존)
* key만 넣고 value를 생략하면 자동으로 'None'이 입력됨

```python
d
```

```python
d['년월'] = '20240924'
d
```

```python
# 기존에 key가 있는 경우 기존 key의 value를 출력하고 새로값은 무시
d.setdefault('년월', 20250924)
```

```python
d
```

```python
d.setdefault('날씨', '맑음')
d
```

```python
for i in range(1,11):
    d.setdefault('숫자', []).append(i)
```

```python
d
```
