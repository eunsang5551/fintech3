```python
# end 값의 기본 \\n
print("헐크", "아이언맨", "스파이더맨")
print()
print("헐크", "아이언맨", "스파이더맨")
```

```python
# end 값의 기본 \\n
print("헐크", "아이언맨", "스파이더맨", end="\n\n")
print("헐크", "아이언맨", "스파이더맨")
```

```python
# end 값의 기본 \\n
print("헐크", "아이언맨", "스파이더맨", end=", ")
print("헐크", "아이언맨", "스파이더맨")
print("헐크", "아이언맨", "스파이더맨")
```

```python
# end 값 \\t  tap키 
print("헐크", "아이언맨", "스파이더맨", end="\t")
print("헐크", "아이언맨", "스파이더맨")
print("헐크", "아이언맨", "스파이더맨")
```

```python
# end 값 \\r  carrige return 커서를 맨 앞으로 이동
print("헐크", "아이언맨", "스파이더맨", end="\r")
print("캐리지리턴 됨")
```
