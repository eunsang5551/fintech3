```python
# for문으로 1-10까지 수가 담긴 리스트 만들기
list1_10 = []
for i in range(1, 11):
    list1_10.append(i)
list1_10
```

```python
# list 표현식으로 1-10까지 수가 담긴 리스트 만들기
list1_10 = [i for i in range(1, 11)] 
list1_10
```

```python
# 리스트 표현식으로 1-10까지 수 중에서 홀수와 짝수를 구분해서 리스트에 담기
result = [f"{i}는 짝수" if i % 2 == 0 else f"{i}는 홀수" for i in range(1,11)]
result
```

# 반복문 + if 문의 조건표현식

```python
# 일반 for if문
evens = []
for i in range(1, 11):
    if i % 2 == 0:
        evens.append(i)
evens
```

```python
# 리스트표현식 if만 있을 때
evens = [i for i in range(1, 11) if i % 2 == 0]
evens
```

```python
# 일반 for if else문
labels = []
for i in range(1, 11):
    if i % 2 == 0:
        labels.append("짝수")
    else:
        labels.append("홀수")
labels
```

# 리스트표현식으로 if else가 포함된 코드만들기

```python
labels = ["짝수" if i % 2 == 0 else "홀수" for i in range(1, 11)]
labels
```

# 리스트와 제네레이터 크기 비교

```python
!pip install pympler
```

```python
import sys
from pympler import asizeof
import psutil, os

# 메모리 사용량 확인 함수
def show_memory(msg):
    process = psutil.Process(os.getpid())
    print(f"{msg}: {process.memory_info().rss / 1024**2:.2f} MB")
    
# 비교 대상 생성
list_var = [x for x in range(1, 1_000_000)]   # 리스트 내포
gen_var  = (x for x in range(1, 1_000_000))   # 제너레이터 표현식

print("==== sys.getsizeof() 결과 ====")
print("리스트:", sys.getsizeof(list_var))
print("제너레이터:", sys.getsizeof(gen_var))

print("\n==== asizeof.asizeof() 결과 (내부 요소 포함) ====")
print("리스트:", asizeof.asizeof(list_var))
print("제너레이터:", asizeof.asizeof(gen_var))

print("\n==== 프로세스 전체 메모리 사용량 비교 ====")
show_memory("리스트 생성 후")
show_memory("제너레이터 생성 후")
```
